```python
import pandas as pd
```

```python
honeyproduction = pd.read_csv('data/honeyproduction.csv')
honeyproduction.head()
```

```python
honeyproduction.shape
```

```python
honeyproduction.columns
```

```python
honeyproduction.info()
```

```python
numericalCol = 0
categoricalCol = 0

numericalColList = []
categoricalColList = []

for col in honeyproduction.columns:
    if honeyproduction[col].dtype == 'object':
        categoricalCol += 1
        categoricalColList.append(col)
    else:
        numericalCol += 1
        numericalColList.append(col)
```

```python
print("Numerical columns = ", numericalCol, " : ", numericalColList)
print("Categorical columns = ", categoricalCol, " : ", categoricalColList)
```

```python
honeyproduction.describe(include='all')
```

```python
honeyproduction[categoricalColList].describe()
```

```python
honeyproduction[numericalColList].describe().T
```

```python
honeyproduction.isnull().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
for nCol in numericalColList:
    print(nCol)
    plt.hist(honeyproduction[nCol])
    plt.xlabel(nCol)
    plt.ylabel("Frequency")
    plt.plot()
    plt.show()
```

```python
for nCol in numericalColList:
    print(nCol)
    sns.histplot(honeyproduction[nCol], kde=True)
    plt.xlabel(nCol)
    plt.ylabel("Frequency")
    plt.plot()
    plt.show()
```

```python
for nCol in numericalColList:
    print(nCol)
    sns.boxplot(x = honeyproduction[nCol])
    plt.xlabel(nCol)
    plt.ylabel("Frequency")
    plt.plot()
    plt.show()
```

```python
for cCol in categoricalColList:
    print(cCol)
    sns.countplot(x = honeyproduction[cCol])
    plt.xlabel(cCol)
    plt.ylabel("Frequency")
    plt.xticks(rotation=45)    
    plt.plot()
    plt.show()
```
