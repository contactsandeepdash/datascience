## Why we need to scale data?

- Input variables may have different units (e.g., feet, KM and Hours)
- Differnces in the scales across input variables may increase difficulty of the problem being modeled
- Some algorithms may benefit from normalizing the input data

### Types of scaling
1. **Normalization**
    - Rescaling of the data from the original range so that all values are within the range of 0 and 1; also known as MinMax scaler
    - accurately estiamte the minimum and maximum observable values
    - formula, y = (x - min) / (max - min)
    - import MinMaxScaler from sklearn.preprocessing for Normalization
2. **Standardization**
    - Standarding a dataset involves rescaling the distribution of values so that the mean of observed values is 0 and the standard deviation is 1
    - This assumes that the observations fit a normal distribution
    - formula, y = (x - mean) / standard_deviation
    - import StandardScaler from sklearn.preprocessing for Standardization
    


## Univariate Analysis
- data has **only 1** varibale
- explores each variable in a dataset, separately
- looks at the range of values as well as central tendency of the values
- describes pattern of response to the vaiable, describing each variable on its own
- Techniques useful for Univariable analysis are:
    - Frequency Distribution Tables
    - Bar charts
    - Histograms
    - Pie Charts
    - Box Plots

Univariate
1. Categorical
    - Numerical EDA -> pd.crosstab
    - Visual EDA -> sns.CountPlot()
2. Continuous
    - Numerical EDA -> dataframe.describe()
    - Visual EDA -> boxplot(), distplot(), kdeplot()

## Bivariate Analysis

- Involves the analysis of 2 variables (often denoted as X, Y) - determining the empirical relationship between them
- Techniques useful for Bivariate analysis are:
    - Scatter plots
    - Correlation Coefficients
    - Chi Square Test

Bivariate
1. Category to category
    - sns.factorPlot()
2. Category to continuous
    - sns.jointPlot()
3. Continuous to category
    - sns.factorPlot().map(sns.kde/dist/box0)

## Multivariate Analysis

- Involves multiple dependent variables resulting in 1 outcome
- Techniques usedful for multivariate analysis are:
    - Principal Component Analysis (PCA)
    - Cluster analysis
    - Multiple Regression Analysis
    - Correlations
